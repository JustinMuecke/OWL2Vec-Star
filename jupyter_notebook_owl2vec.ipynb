## Running OWL2Vec*

```python
from owl2vec_star import owl2vec_star
import os
import numpy as np
from tqdm import tqdm
from random import shuffle
from owlready2 import get_ontology
import traceback
```

```python
def _create_graph_embedding(embeddings):
    words = embeddings.key_to_index
    vectors = [embeddings[word] for word in words]
    graph_embedding = np.mean(vectors, axis=0)
    return graph_embedding
```

```python
def _get_embeddings(directory_path, number, consistency):
    rows = []

    filenames = os.listdir(directory_path)
    shuffle(filenames)

    for i in tqdm(range(number)):
        try:
            gensim_model = owl2vec_star.extract_owl2vec_model(f"{directory_path}{filenames[i]}", "./default.cfg", True, True, True)
            graph_embedding = _create_graph_embedding(gensim_model.wv)
            print("GOT EMBEDDING")
            rows.append([filenames[i], graph_embedding, consistency])
        except Exception as exception:
            traceback.print_exc()
            continue
        
    return rows
```

```python
inconsistent_data = _get_embeddings("../data/ont_modules_inconsistent_rdf/", 1, False)
```
